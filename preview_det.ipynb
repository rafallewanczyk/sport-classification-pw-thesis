```python
from preprocessing.video import Video
from preprocessing.dto.sequence_dto import SequenceDto
from preprocessing.inference_service.external_model.yolov7_pose_tracker.utils.plots import plot_one_box_kpt, \
    plot_skeleton_kpts
from pathlib import Path
import cv2
import numpy as np

v = Video(Path('/home/rafa/SVW/Videos/tennis/145___14d52737e37b4fbe8b4aadf4fecaa78a.mp4'))
detections = v.build_detections()
detections.head()
```

```python
cap = cv2.VideoCapture(v.video_reshaped_path.as_posix())
video = [np.zeros((1080, 1920, 3))]
while (cap.isOpened()):
    ret, img = cap.read()
    if not ret:
        break
    video.append(img)
cap.release()

```

```python

from preprocessing.dto.detection_dto import DetectionDto

persons = detections[DetectionDto.PERSON_ID].unique()
frames = detections[DetectionDto.FRAME_ID].unique()
for frame in frames:
    # frame = np.zeros
    skeletons = detections[detections[DetectionDto.FRAME_ID] ==frame]["skeleton"]
    bboxes = detections[detections[DetectionDto.FRAME_ID] ==frame]["bbox"]
    colors = detections[detections[DetectionDto.FRAME_ID] ==frame]["unique_color"]
    frame_img = video[frame]
    for bbox, skeleton, color in zip(bboxes, skeletons, colors):
        accs = np.average(skeleton[2::3])
        if accs < 0.5:
            continue
        x = [bbox[0], bbox[1], bbox[0] + bbox[2], bbox[1] + bbox[3]]
        plot_one_box_kpt(x=x, im=frame_img, kpt_label=True, kpts=skeleton, steps=3, color=color, line_thickness=10)
        print(skeleton.reshape((-1, 3)))
    cv2.imshow('frame', frame_img)
    cv2.waitKey(0)
cv2.destroyAllWindows()
```
